### Atrinkite visus skirtingus miestus iš kurių yra šie respondentai.

```python
df['Miestas.1'] = df['Miestas.1'].str.title().str.strip()

miestai = df['Miestas.1'].unique()

print(miestai)
```

### Raskite didžiausią ir mažiausią įvestus atlyginimus.

```python
didziausias_atlyginimas = df['Atlyginimas'].max()
maziausias_atlyginimas = df['Atlyginimas'].min()

print(didziausias_atlyginimas)
print(maziausias_atlyginimas)

```

### Raskite atlyginimų vidurkį.

```python
vidurkis = round(sum(df['Atlyginimas']) / len(df['Atlyginimas']))

print(vidurkis)
```

### Raskite atlyginimų vidurkius pagal lygį (1 - junior, 2 - mid, 3 - senior).

```python
atlyginimu_vidurkiai_pagal_lygi = df.groupby('Lygis')['Atlyginimas'].mean()
print(round(atlyginimu_vidurkiai_pagal_lygi))
```
